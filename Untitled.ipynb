```python
import uproot
import awkward as ak
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mplhep as hep
plt.style.use(hep.style.ROOT)
```

```python
from coffea.nanoevents.methods import vector
ak.behavior.update(vector.behavior)
```

```python
import boost_histogram as bh
```

```python
keys = ['NoSDiagram','NoOS','Incl','DiProd']
```

```python
hists = {}
```

```python
nbin = 20
mins = {'eta':-4,'pt':0}
maxs = {'eta':4,'pt':2000}
```

```python
for key in keys:
    infile = uproot.concatenate(f"{key}.root:mytree")
    particle = ak.zip({
    'x':infile['px'],
    'y':infile['py'],
    'z':infile['pz'],
    't':infile['E'],
    'id':infile['pdgID'],
    'status':infile['pdgStatus'],
    },with_name='LorentzVector')
    
    tau_mask = abs(particle.id) == 15
    tau = particle[tau_mask]
    for kin,kin_tau in zip(['eta','pt'],[tau.eta,tau.pt]):
        
        hists[key+'-tau-'+kin] = bh.Histogram(bh.axis.Regular(nbin,mins[kin],maxs[kin]), storage=bh.storage.Weight())
        hists[key+'-tau-'+kin].fill(kin_tau[:,0])
    mu_mask = np.logical_and(abs(particle.id) == 13,particle.status!=-1)
    mu = particle[mu_mask]
    for kin,kin_tau in zip(['eta','pt'],[mu.eta,mu.pt]):
        
        hists[key+'-mu-'+kin] = bh.Histogram(bh.axis.Regular(nbin,mins[kin],maxs[kin]), storage=bh.storage.Weight())
        hists[key+'-mu-'+kin].fill(kin_tau[:,0])
    b_mask = abs(particle.id) == 5
    b= particle[b_mask]
    for kin,kin_tau in zip(['eta','pt'],[b.eta,b.pt]):
        
        hists[key+'-b-'+kin] = bh.Histogram(bh.axis.Regular(nbin,mins[kin],maxs[kin]), storage=bh.storage.Weight())
        hists[key+'-b-'+kin].fill(kin_tau[:,0])
```

```python
for kin in ['eta','pt']:
# for key in keys:
    for part in ['tau','mu','b']:
        plt.figure()
        for key in keys:
            tmpkey = key+f'-{part}-'+kin
            hep.histplot(hists[tmpkey],label=key)
            plt.legend()
            plt.xlabel(kin+'_'+part)
            plt.ylabel('A.U.')
```
