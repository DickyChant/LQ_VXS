```python
import uproot
import awkward as ak
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mplhep as hep
plt.style.use(hep.style.ROOT)
```

```python
from coffea.nanoevents.methods import vector
ak.behavior.update(vector.behavior)
```

```python
import boost_histogram as bh
```

```python
keys = ['NoSDiagram','NoOS','Incl','DiProd']
```

```python
hists = {}
```

```python
nbin = 20
mins = {'eta':-4,'pt':0}
maxs = {'eta':4,'pt':2000}
```

```python
for key in keys:
    infile = uproot.concatenate(f"{key}.root:mytree")
    particle = ak.zip({
    'x':infile['px'],
    'y':infile['py'],
    'z':infile['pz'],
    't':infile['E'],
    'id':infile['pdgID'],
    'status':infile['pdgStatus'],
    },with_name='LorentzVector')
    
    tau_mask = abs(particle.id) == 15
    tau = particle[tau_mask][:,0]
    for kin,kin_obj in zip(['eta','pt'],[tau.eta,tau.pt]):
        
        hists[key+'-tau-'+kin] = bh.Histogram(bh.axis.Regular(nbin,mins[kin],maxs[kin]), storage=bh.storage.Weight())
        hists[key+'-tau-'+kin].fill(kin_obj)
    mu_mask = np.logical_and(abs(particle.id) == 13,particle.status!=-1)
    mu = particle[mu_mask][:,0]
    for kin,kin_obj in zip(['eta','pt'],[mu.eta,mu.pt]):
        
        hists[key+'-mu-'+kin] = bh.Histogram(bh.axis.Regular(nbin,mins[kin],maxs[kin]), storage=bh.storage.Weight())
        hists[key+'-mu-'+kin].fill(kin_obj)
    b_mask = abs(particle.id) == 5
    b= particle[b_mask]
    b1 = b[:,0]
    b2 = b[:,1]
    for kin,kin_obj in zip(['eta','pt'],[b1.eta,b1.pt]):
        hists[key+'-b1-'+kin] = bh.Histogram(bh.axis.Regular(nbin,mins[kin],maxs[kin]), storage=bh.storage.Weight())
        hists[key+'-b1-'+kin].fill(kin_obj)
    for kin,kin_obj in zip(['eta','pt'],[b2.eta,b2.pt]):
        hists[key+'-b2-'+kin] = bh.Histogram(bh.axis.Regular(nbin,mins[kin],maxs[kin]), storage=bh.storage.Weight())
        hists[key+'-b2-'+kin].fill(kin_obj)
    mub1 = mu + b1
    hists[key+'-m_mb1'] = bh.Histogram(bh.axis.Regular(nbin,901,1101), storage=bh.storage.Weight())
    hists[key+'-m_mb1'].fill(mub1.mass)
    
    eta_mub1 = np.abs(mu.eta - b1.eta)
    hists[key+'-eta_mb1'] = bh.Histogram(bh.axis.Regular(nbin,0,4), storage=bh.storage.Weight())
    hists[key+'-eta_mb1'].fill(eta_mub1)
    
    mub2 = mu + b2
    hists[key+'-m_mb2'] = bh.Histogram(bh.axis.Regular(nbin,901,1101), storage=bh.storage.Weight())
    hists[key+'-m_mb2'].fill(mub2.mass)
    
    eta_mub2 = np.abs(mu.eta - b2.eta)
    hists[key+'-eta_mb2'] = bh.Histogram(bh.axis.Regular(nbin,0,4), storage=bh.storage.Weight())
    hists[key+'-eta_mb2'].fill(eta_mub2)
    
    taub1 = tau + b1
    hists[key+'-m_tab1'] = bh.Histogram(bh.axis.Regular(nbin,901,1101), storage=bh.storage.Weight())
    hists[key+'-m_tab1'].fill(mub1.mass)
    
    eta_taub1 = np.abs(tau.eta - b1.eta)
    hists[key+'-eta_tab1'] = bh.Histogram(bh.axis.Regular(nbin,0,4), storage=bh.storage.Weight())
    hists[key+'-eta_tab1'].fill(eta_taub1)
    
    taub2 = tau + b2
    hists[key+'-m_tab2'] = bh.Histogram(bh.axis.Regular(nbin,901,1101), storage=bh.storage.Weight())
    hists[key+'-m_tab2'].fill(mub2.mass)
    
    eta_taub2 = np.abs(tau.eta - b2.eta)
    hists[key+'-eta_tab2'] = bh.Histogram(bh.axis.Regular(nbin,0,4), storage=bh.storage.Weight())
    hists[key+'-eta_tab2'].fill(eta_taub2)
    
```

```python
for kin in ['eta','pt']:
# for key in keys:
    for part in ['tau','mu','b1','b2']:
        plt.figure()
        for key in keys:
            tmpkey = key+f'-{part}-'+kin
            hep.histplot(hists[tmpkey],label=key)
            plt.legend()
            plt.xlabel(kin+'_'+part)
            plt.ylabel('A.U.')
```
